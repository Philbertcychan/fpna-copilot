```python
import pandas as pd
df = pd.read_csv("../data/fake_fpna.csv")
print(df.head())
print(df.describe())
```

```python
# Revenue variance
df['Revenue_Variance'] = df['Revenue'] - df['Budget_Revenue']
df['Revenue_Variance_Pct'] = (df['Revenue_Variance'] / df['Budget_Revenue']) * 100

# Expenses variance
df['Expenses_Variance'] = df['Expenses'] - df['Budget_Expenses']
df['Expenses_Variance_Pct'] = (df['Expenses_Variance'] / df['Budget_Expenses']) * 100

# Check
print(df[['Month', 'Revenue', 'Budget_Revenue', 'Revenue_Variance', 'Revenue_Variance_Pct',
          'Expenses', 'Budget_Expenses', 'Expenses_Variance', 'Expenses_Variance_Pct']])
```

```python
# Save summary CSV for AI or dashboard
df.to_csv("../outputs/fake_fpna_summary.csv", index=False)
```
